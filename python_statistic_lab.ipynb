```python
import pandas as pd
#Загрузим файл
df = pd.read_csv('/Users/peter/Downloads/auto.csv')
df.head(20)
```

```python
#Рассчитаем основные статистики для метрических переменных
df.describe()
```

```python
df.info()
```

```python
#мы не считаем корреляцию для номинативных переменных, поэтому оставим только метрические переменные для нормирования данных
spisok = []
for col in df.columns:
    if df[col].dtype != 'object':
        spisok.append(col)
df = df.filter(items = spisok)
df.head()
```

```python
#проведем нормирование данных
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sdf = pd.DataFrame(scaler.fit_transform(df))
#вернем колонкам оригинальные названия
sdf.columns = spisok
sdf.head()
```

```python
#построим матрицу корреляций r pirson
sdf.corr()
```

```python
#визуализируем матрицу корреляций
import seaborn as sns
sns.pairplot(sdf)
```

```python
#Визуализировать распределение переменной ‘Price’,рассчитать показатели skewness и kurtosis
```

```python
sns.distplot(sdf['price'])
```

```python
from scipy import stats
print('kurtosis =', stats.describe(sdf['price'])[-1])
print('skewness =', stats.describe(sdf['price'])[-2])
```

```python
#5.Визуализировать несколько взаимосвязей междупеременными
```

```python
#Как мы знаем, по корреляции нельзя судить о причинно следственной связи, но мы видим что с увеличением объема двигателя, 
#стоимость автомобиля растет. 
#Есть сильная линейная взаимосвязь
```

```python
#Видим несколько интересных корреляций:
#price и engine size
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
corr, p_value = pearsonr(sdf['price'], df['engine-size'])
plt.scatter(sdf['price'], df['engine-size'], marker='.')
print(corr)
```

```python
#Между длинной автомобиля и весьм карбюратора так же наблюдается сильная положительная взаимосвязь.
#С возрастанием длины автомобиля, растет весь карбюратора.
```

```python
#curb-weight и length
corr, p_value = pearsonr(sdf['curb-weight'], df['length'])
plt.scatter(sdf['length'], df['curb-weight'], marker='.')
print(corr)
```

```python
#Выдвинуть несколько гипотез по набору имеющихсяпеременных и проверить эти гипотезы с использованием t-test.
#Проверм отличается ли средняя стоимость у автомобилей ауди и бмв по нашей выборке
# p-value > 5%, похоже что нет
stats.ttest_ind(df[ df['make'] == 'bmw' ]['price'].sample(n=5), df[ df['make'] == 'audi' ]['price'].sample(n=5))
```
